# Intro to Tensors
A tensor is a multi-dimensional array which can run on a GPU. They are a generalization of matrices to an arbitrary number of dimensions.

```python
# Create a scalar (0-dimensional tensor)
scalar = torch.tensor(7)
print(scalar)
```

```python
# Get the number of dimensions of the scalar
print(scalar.ndim)

# Get the number of elements in the scalar
print(scalar.item())

# Get the shape of the scalar
print(scalar.shape)
```

Vector is a 1-dimensional tensor with more than one element.

```python
vector = torch.tensor([7, 7])
print(vector)
print(vector.ndim)
```

```python
vector.shape
```

Matrix is a 2-dimensional tensor.

```python
MATRIX = torch.tensor([[7, 8], [9, 10]])

print(MATRIX)
```

```python
MATRIX.ndim
```

```python
MATRIX.shape
```

# RANDOM TENSORS
Random tensors are tensors that are filled with random numbers. They are often used to initialize the weights of a neural network before training. So, after training, the random numbers will be updated to values that better represent the patterns in the data.

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
```

```python
print(random_tensor)
```

```python
random_tensor.ndim
```

```python
random_tensor.shape
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_tensor = torch.rand(size=(224, 224, 3))  # height, width, color channels
```

```python
random_image_tensor.ndim, random_image_tensor.shape
```

```python
# Zeros and Ones
# Create a tensor of all zeros or ones. This is useful when you want to create a tensor of a specific size but don't care about the values (e.g. for embedding layers, padding, etc.)

zeros = torch.zeros(size=(3, 4))
ones = torch.ones(size=(3, 4))

print(zeros)
print(ones)
```

```python
zeros.dtype, ones.dtype
```

# Tensor Data Types

**Note:** The biggest erros when working with PyTorch are:
- 1. Tensor with incorrect data type,
- 2. Tensors with incorrect shape,
- 3. Tensors with incorrect device.

```python
float_tensor = torch.tensor(
    [3.0, 6.0, 9.0],
    dtype=torch.float16,  # what dtype is the tensor
    device=None,  # what device is the tensor on (GPU or CPU)
    requires_grad=False,  # whether or not to track gradients with this tensor. This is important for machine learning, if you set requires_grad=True, PyTorch will keep track of all operations on the tensor
    # and allow you to calculate gradients for optimization.
)
```
